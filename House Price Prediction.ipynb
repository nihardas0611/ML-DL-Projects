```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import sklearn.datasets
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#Importing House Price Dataset
house_price_dataset=sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
house_price_dataframe=pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
house_price_dataframe['Price']=house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.describe()
```

```python
#Understanding the correlation
correlation=house_price_dataframe.corr()
```

```python
#Constructing a heat map
plt.figure(figsize=(8,8))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':10},cmap='Reds')
```

```python
#Splitting the data and the target
X=house_price_dataframe.drop(columns='Price',axis=1)
Y=house_price_dataframe['Price']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
#loading the model
model=XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
#Evaluation
training_data_prediction=model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R sqaured error
score_1=metrics.r2_score(Y_train,training_data_prediction)

#Mean Absolute Error
score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)

print("R squared error: ",score_1)
print("Mean Absolute Error: ",score_2)
```

Visualize the actual prices vs predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual price vs Predicted price")
plt.show()
```

```python
test_data_prediction=model.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
# R sqaured error
score_1=metrics.r2_score(Y_test,test_data_prediction)

#Mean Absolute Error
score_2=metrics.mean_absolute_error(Y_test,test_data_prediction)

print("R squared error: ",score_1)
print("Mean Absolute Error: ",score_2)
```

Visualize the actual prices vs predicted prices

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual price vs Predicted price")
plt.show()
```
