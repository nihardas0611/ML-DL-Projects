This Dataset is Highly Unbalanced

0-->Normal Transaction

1-->Fraudulent Transaction

```python
#Separating the data for analysis
legit=credit_dataset[credit_dataset.Class==0]
fraud=credit_dataset[credit_dataset.Class==1]
```

```python
print(legit.shape,fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values for both transaction
credit_dataset.groupby('Class').mean()
```

```python
#Under Sampling
#Build a sample dataset containing similar distribution of normal and fraudulent transactions
legit_sample=legit.sample(n=492)
```

```python
#Concatenating two dataset
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.15,stratify=Y,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
training_data_prediction=accuracy_score(X_train_prediction,Y_train)
print(training_data_prediction)
```

```python
X_test_prediction=model.predict(X_test)
test_data_prediction=accuracy_score(X_test_prediction,Y_test)
print(test_data_prediction)
```

```python
#Predictive System
input_data=#(Provide Data)

input_data_as_nparray=np.asarray(input_data)
reshaped_array=input_data_as_nparray.reshape(1,-1)

prediction=model.predict(reshaped_array)
print(prediction)

if(prediction[0]==0):
  print('Legit Transaction')
else:
  print('Fraudulent Transaction')
```
